# Multi Class Classification

```python
import tensorflow as tf
```

```python
from tensorflow.keras.datasets import fashion_mnist
```

```python
(x_train,y_train),(x_test,y_test) = fashion_mnist.load_data()
```

```python
x_train[0].shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(x_train[0])
```

```python
y_train[0]
```

```python
class_name=["T-shirt/top","Trouser","Pullover",
            "Dress","Coat","Sandal","Shirt",
            "Sneaker","Bag","Ankle boot"]
```

```python
import random
```

```python
n=9
for i in range(n):
    plt.subplot(3,3,i+1)
    r=random.choice(range(len(x_train)))
    plt.imshow(x_train[r],cmap=plt.cm.gray)
    plt.title(class_name[y_train[r]])
```

```python
len(x_train)
```

```python
x_train[0].shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=x_train[0].shape))
model.add(tf.keras.layers.Dense(4,activation="relu"))
model.add(tf.keras.layers.Dense(4,activation="relu"))
model.add(tf.keras.layers.Dense(10,activation="softmax"))

model.compile(
    #loss = tf.keras.losses.CategoricalCrossentropy(),
    loss = tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics="accuracy"
)

hist = model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=x_train[0].shape))
model.add(tf.keras.layers.Dense(10,activation="relu"))
model.add(tf.keras.layers.Dense(10,activation="softmax"))

model.compile(
    loss =tf.keras.losses.CategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = "accuracy"
)

hist =model.fit(x_train,tf.one_hot(y_train,depth=10),epochs=10)
```

```python
import pandas as pd
```

```python
pd.DataFrame(hist.history).plot(title="Normalized data")
```

```python
y_pred = model.predict(x_test)
```
